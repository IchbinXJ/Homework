```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./stockholm.csv')
df.head()
```

```python
for i in range(5):
    for j in range(df.shape[0]):
        value = df.iloc[j,i]
        value_str = str(value)
        if 'a' in value_str.lower() or 'b' in value_str.lower():
            print("in Column {} Roll {} locate {}".format(i, j, value)) #collect and printout
            value_str = value_str.replace('a', '').replace('b', '')
            df.iloc[j,i] = float(value_str)   #data clean  
```

```python
##########################Task 2


df['average'] = (pd.to_numeric(df['T_max']) + pd.to_numeric(df['T_min'])) / 2 # average the data and create a new column
df.to_csv('./stockholm_temperature.csv')
```

```python
#############Task 3

def Kelv(a):
    return a+273.15


df['to_Kelv'] = df['average'].apply(lambda average: Kelv(average))
```

```python
df
```

```python
####################Task 4
max_value = df['average'].max()
min_value = df['average'].min()

max_value_k = Kelv(max_value)
min_value_k = Kelv(min_value)

print("max and min temperature are {} and {}, to Kelvin are {} K and {} K ".format(max_value, min_value, max_value_k, min_value_k))
```
